# Diabetes Prediction using Random Forest Classifier

This notebook demonstrates a complete machine learning pipeline for diabetes prediction using the Random Forest algorithm. The workflow includes data loading, preprocessing, model training, and evaluation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:, df.columns != 'Outcome']
y = df.iloc[:, df.columns == 'Outcome']
```

```python
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(xtrain, ytrain.values.ravel())
```

```python
output = model.predict(xtest)
output
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(output, ytest)
print("The accuracy score is:", acc)
```
