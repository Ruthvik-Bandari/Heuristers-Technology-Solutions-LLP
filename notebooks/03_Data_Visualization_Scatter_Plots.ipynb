# Data Visualization - Scatter PlotsCreating scatter plots and combined visualizations using Matplotlib for exploratory data analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('TASK 8.csv')
```

```python
df
```

```python
plt.bar(x=df['FORUM'],height=df['VALUES'],color='red')
```

```python
df.plot.scatter(x='FORUM',y='VALUES')
```

```python
marks=np.array([45,30,50,54,70,44,43])
rollno=np.array([23,45,66,56,33,48,75])
plt.scatter(rollno,marks)
plt.xlabel('rollno')
plt.ylabel('marks')
plt.plot(rollno,marks,color='red')
plt.show()
```
