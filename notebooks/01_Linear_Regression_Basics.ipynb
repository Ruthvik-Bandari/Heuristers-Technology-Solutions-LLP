# Linear Regression BasicsIntroduction to Linear Regression using scikit-learn. This notebook demonstrates fitting a linear model to sample data and making predictions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([2,11,15,50,70,81,90,115,125,145])
y=np.array([4,19,40,60,100,140,150,210,240,400])
```

```python
x=x.reshape(-1,1)
linreg=LinearRegression()
```

```python
linreg.fit(x,y)
```

```python
y_pre =linreg.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y,color='pink')
plt.show
```
